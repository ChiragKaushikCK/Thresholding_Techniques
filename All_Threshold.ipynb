```python
import cv2
import matplotlib.pyplot as plt

#load the image
image=cv2.imread('/content/Sample_Img.webp',cv2.IMREAD_GRAYSCALE)

#apply global thresholding

ret, threshold=cv2.threshold(image,127,255,cv2.THRESH_BINARY)

#save the threshold image
cv2.imwrite('threshold_image.jpg',threshold)

#display original and threshold image
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(image,cmap='gray')
plt.title('Original Image')

plt.subplot(1,2,2)
plt.imshow(threshold,cmap='gray')
plt.title('Threshold Image')
plt.show()
```

```python
#using for loop for changing threshold value

import cv2
import matplotlib.pyplot as plt

#load the image
image=cv2.imread('/content/Sample_Img.webp',cv2.IMREAD_GRAYSCALE)

#apply global thresholding
for i in range(100,200,20):
  ret, threshold=cv2.threshold(image,i,255,cv2.THRESH_BINARY)

  #save the threshold image
  cv2.imwrite('threshold_image.jpg',threshold)

  #display original and threshold image
  plt.figure(figsize=(10,5))

  plt.subplot(1,2,1)
  plt.imshow(image,cmap='gray')
  plt.title('Original Image')

  plt.subplot(1,2,2)
  plt.imshow(threshold,cmap='gray')
  plt.title('Threshold Image')
  plt.show()
```

**Adaptive Thresholding**

```python
import cv2
from google.colab.patches import cv2_imshow


```

```python
image = cv2.imread('/content/Sample_Img.webp',cv2.IMREAD_GRAYSCALE)

thresh_mean = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,5)


cv2_imshow(image)
cv2_imshow(thresh_mean)
```

**Gaussian Threshold**

```python
gaus_thresh = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,-5)

cv2_imshow(image)
cv2_imshow(gaus_thresh)
```
